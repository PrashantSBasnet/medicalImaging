```python
import os
import pandas as pd
```

```python
base_url = r'E:\DataScience Projects\prashant\dataset'
df = os.path.join(base_url, 'train.csv')
```

```python
train_df =  pd.read_csv(df)
```

```python
train_df
```

```python
na_counts = train_df['Pneumonia'].isna().sum()
print(na_counts)
```

```python
# Step 1: Remove rows with NaN in the Pneumonia column
df_non_nan_pneumonia = train_df[train_df['Pneumonia'].notna()]

# Step 2: Select only specific columns
# Adjust the columns below based on the ones you want to retain
selected_columns = ['Path', 'Sex', 'Age', 'Frontal/Lateral', 'AP/PA', 'Pneumonia']
filtered_df = df_non_nan_pneumonia[selected_columns]
```

```python
filtered_df
```

```python
train_df['Pneumonia'].count
```

```python
import matplotlib.pyplot as plt



pneumonia_labels = {1.0: 'Positive', 0.0: 'Absent', -1.0: 'Uncertain'}
filtered_df['Pneumonia_Labeled'] = filtered_df['Pneumonia'].replace(pneumonia_labels)

# Step 3: Count each label in the labeled column
pneumonia_counts = filtered_df['Pneumonia_Labeled'].value_counts()

# Step 4: Plot the counts
plt.figure(figsize=(8, 6))
pneumonia_counts.plot(kind='bar')
plt.title('Distribution of Pneumonia Labels')
plt.xlabel('Pneumonia Label')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```
